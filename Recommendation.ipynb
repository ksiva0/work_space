#### Recommend based on Grade

```python
def recommend_assessment_by_grade(grade, top_n=5):
    grade_data = df[df['GRADE'] == grade].reset_index(drop=True)  # Filter the data on grade
    
    # Check if the grade has enough to make recommendations
    if grade_data.shape[0] <= 1:
        print("Not enough data to provide recommendations for this grade.")
        return
    
    features = grade_data[['overallScore', 'questionsDone', 'totalNumberQuestions']]
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    similarity_matrix = cosine_similarity(features_scaled)  # Calculate similarity matrix for assessments

    recommendations = []
    for idx in range(len(grade_data)):
        similar_indices = np.argsort(-similarity_matrix[idx])[:top_n+1]
        similar_indices = similar_indices[similar_indices != idx]  # Exclude the assessment itself
        recommended_assessments = grade_data.iloc[similar_indices][['studentName', 'assessmentTitle', 'overallScore', 'GRADE']]
        recommendations.append((grade_data.iloc[idx]['assessmentTitle'], recommended_assessments))
    
    return recommendations
```
